```python
import polars as pl
import numpy as np
from pathlib import Path
import plotly.express as px
project_dir = Path(globals()['_dh'][0]).parent

```

```python

df = pl.read_csv(project_dir/"data"/"titanic.csv")
df.head(5)
```

```python
px.scatter( y=df.select("Pclass").to_series(), x=df.select("Age").to_series(), color=df.select("Survived").to_series(), labels=["class", "Age"], title="Titanic")
```

Two pie charts: One that shows of the survivors, which class they were in, the second that shows of the non-survivors, which class they were in.

```python
px.pie(names=df.filter(pl.col("Survived")==1).select("Pclass").to_series(), title="Classes of Survivors")
```

```python
px.pie(names=df.filter(pl.col("Survived")==0).select("Pclass").to_series(), title="Classes of !Survivors")
```

A 3 axis parallel coordinates plot of Survived, Class, Age in that order.

```python
px.parallel_coordinates(df.select(["Survived","Pclass", "Age"]).to_pandas())
```
